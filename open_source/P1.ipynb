### Feature Missingness and Interpretation

| Feature | Missing Rate | Processing Logic |
|--------|--------------|----------------|
| **Urosurgeries, Gynosurgeries** | ~99% missing | Missing implies “no surgery history”; the feature provides no learnable signal → **drop column** |
| **BMI, Age** | ~91% missing | Imputation would introduce artificial correlations → **drop column** |
| **Medvits (Medication/Supplements)** | ~91% missing | If unique values are clear → **classify**; if medication types are inconsistent → **drop column** |
| **Miscarriages, Abortions, Numberpreg, Livingkids, Breastfeeding** | ~91% missing | Missing generally indicates “no such event” → **impute 0** |
| **UnusualBleeding, LengthofMenses** | Very low missingness | **Retain** (reliable) |
| **ReproductiveCategory, CycleWithPeakorNot, LengthofCycle** | No missingness | **Core stable features — retain** |

---

### Final Processing Strategy (Concise)

| Feature Category | Processing Method |
|-----------------|------------------|
| **Surgery history (Urosurgeries, Gynosurgeries)** | **Drop columns** |
| **Physiological traits (BMI, Age)** | **Drop columns** |
| **Medication/Supplements (Medvits)** | **Classify if categories clear; otherwise drop** |
| **Reproductive / Pregnancy / Breastfeeding history** | **Impute missing with 0** |
| **Cycle behavior & symptom features** | **Retain all — core modeling features** |
| **LengthofCycle** | **Use as target variable (y)** |

```python
import re
```

```python
# Keywords for classification by physiological impact level

# 3 = Strong direct hormonal/endocrine impact (HPO axis / ovulation regulation)
kw_strong = [
    'progesterone', 'prometrium', 'medroxy',
    'levothyroxine', 'synthroid', 'levoxyl', 'armour thyroid', 'levothroid',
    'clomid', 'letrozole', 'thyroid'
]

# 2 = Moderate systemic effect (mood, metabolism, inflammation, autonomic, etc.)
kw_moderate = [
    # Antidepressants / anxiolytics
    'wellbutrin', 'citalopram', 'celexa', 'effexor', 'paxil', 'lexapro', 'zoloft', 'fluoxetine', 'cymbalta',
    # ADHD stimulants
    'adderall', 'adderal', 'adderall xr',
    # Antipsychotic / neuropathic pain
    'seroquel', 'neurontin',
    # Metabolic / insulin-sensitising
    'metformin',
    # Allergy / inflammation / respiratory
    'singulair', 'zyrtec', 'claritin', 'advair', 'albuterol', 'pulmacort',
    # Cardiovascular / autonomic
    'hydrochlorothiazide', 'toprol xl', 'provigil', 'ambien', 'maxalt',
    # Diagnostic / condition markers
    'fibromyalgia', 'hypothyroidism', 'insulin', 'migraine', 'minocycline', 'humalog'
]

# 1 = Nutritional supplements (low direct hormonal relevance)
kw_supp = [
    'multivit', 'vitamin', 'folic', 'fish oil', 'omega', 'probiotic',
    'calcium', 'magnesium', 'centrum', 'optivite', 'prenatal', 'one-a-day', 'krill oil', 'iron supplement'
]

def mk_pattern(words):
    """Build case-insensitive regex pattern for list of keywords."""
    return re.compile('|'.join(re.escape(w) for w in words), flags=re.IGNORECASE) if words else None

pat_strong   = mk_pattern(kw_strong)
pat_moderate = mk_pattern(kw_moderate)
pat_supp     = mk_pattern(kw_supp)

txt = df['Medvitexplain']

# Medication classification labels:
# 3 = Strong endocrine effect
# 2 = Moderate systemic effect
# 1 = Nutritional supplement
# 0 = No identifiable medication
Med_Class = pd.Series(0, index=df.index, dtype='int8')

# Assign classes in ascending priority, then override:
if pat_supp:
    Med_Class = Med_Class.mask(txt.str.contains(pat_supp, na=False), 1)
if pat_moderate:
    Med_Class = Med_Class.mask(txt.str.contains(pat_moderate, na=False), 2)
if pat_strong:
    Med_Class = Med_Class.mask(txt.str.contains(pat_strong, na=False), 3)

df['Med_Class'] = Med_Class

# Summary
print("Med_Class distribution:")
print(df['Med_Class'].value_counts(dropna=False).sort_index())

print("\nRows per class (for manual sanity check):")
for c in [1, 2, 3]:
    ex = df.loc[df['Med_Class'] == c, 'Medvitexplain']
    print(f"\n== Class {c} ==")
    print(ex.to_string(index=False))
```

```python
def fill_by_client_mean(df, col, id_col="ClientID"):
    """
    Fill missing values in `col` using the mean of other records
    from the same client. Rows that cannot be filled are removed.
    """
    missing_rows = df[df[col].isna()]
    rows_to_drop = []
    filled_records = []

    for idx, row in missing_rows.iterrows():
        client = row[id_col]
        
        # Values from same client that are valid
        client_values = df[(df[id_col] == client) & (df[col].notna())][col]
        
        if len(client_values) > 0:
            new_value = int(round(client_values.mean()))
            df.at[idx, col] = new_value
            filled_records.append((idx, client, new_value))
        else:
            rows_to_drop.append(idx)

    df = df.drop(index=rows_to_drop)
    print(f"[{col}] Filled: {len(filled_records)}, Removed: {len(rows_to_drop)}, Remaining missing: {df[col].isna().sum()}")
    
    return df, filled_records


# Apply to both variables
df_new, filled_ov = fill_by_client_mean(df_new, "EstimatedDayofOvulation")
df_new, filled_len = fill_by_client_mean(df_new, "LengthofMenses")

df_new.head()
```

# Training

## GroupKFold
Apply GroupKFold cross-validation to ensure that all cycles belonging to the same individual are assigned to the same fold. This prevents information leakage and artificially inflated performance estimates due to within-subject autocorrelation.

```python
from sklearn.model_selection import GroupKFold

# Save cleaned dataset
df_new.to_csv("/Users/xujing/FYP/reproduction/dataset/menstrual-cycle-data-kaggle/cleaned_menstrual_data.csv", index=False)

# Feature set and target
features = [
    'Med_Class', 'ReproductiveCategory', 'Numberpreg', 'Miscarriages', 'Abortions',
    'Breastfeeding', 'CycleWithPeakorNot', 'UnusualBleeding',
    'LengthofMenses', 'EstimatedDayofOvulation'
]

X = df_new[features]
y = df_new['LengthofCycle']
groups = df_new['ClientID']  # Ensure cycles from the same individual stay in the same fold

# Grouped cross-validation (prevents data leakage across individuals)
gkf = GroupKFold(n_splits=10)
folds = list(gkf.split(X, y, groups=groups))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np
```

```python
# Evaluate Decision Tree using GroupKFold
THRESHOLD_DAYS = 1.0  # 认为“预测精准”的阈值（天），需要可改就改这里

dt_mae_scores = []
dt_r2_scores = []
dt_max_errors = []
dt_within_counts = []
dt_counts = []

all_abs_errors = []

for fold, (train_idx, test_idx) in enumerate(gkf.split(X, y, groups=groups), 1):

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    dt = DecisionTreeRegressor(max_depth=5, random_state=42)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # 新增：误差统计
    abs_errors = np.abs(y_test - y_pred)
    max_err = abs_errors.max()
    within = (abs_errors <= THRESHOLD_DAYS).sum()
    cnt = abs_errors.shape[0]

    dt_mae_scores.append(mae)
    dt_r2_scores.append(r2)
    dt_max_errors.append(max_err)
    dt_within_counts.append(within)
    dt_counts.append(cnt)

    all_abs_errors.extend(abs_errors.tolist())

    print(f"Fold {fold}: MAE={mae:.3f}, R²={r2:.3f}, "
          f"MaxErr={max_err:.3f} days, "
          f"Precise(≤{THRESHOLD_DAYS:.1f}d)={within}/{cnt} ({within/cnt*100:.1f}%)")

print("\n=== Decision Tree 10-Fold CV Results ===")
print(f"Mean MAE: {np.mean(dt_mae_scores):.3f} ± {np.std(dt_mae_scores):.3f}")
print(f"Mean R²: {np.mean(dt_r2_scores):.3f} ± {np.std(dt_r2_scores):.3f}")
print(f"Mean MaxErr per fold: {np.mean(dt_max_errors):.3f} ± {np.std(dt_max_errors):.3f} days")

# 全局统计：所有测试样本上的最大误差与精准命中数
all_abs_errors = np.array(all_abs_errors)
overall_max_err = all_abs_errors.max()
overall_within = (all_abs_errors <= THRESHOLD_DAYS).sum()
overall_cnt = all_abs_errors.size

# ✅ 新增：误差 > 10 天的样本数量
large_err_count = (all_abs_errors > 5).sum()
large_err_ratio = large_err_count / overall_cnt * 100

print(f"Overall MaxErr: {overall_max_err:.3f} days")
print(f"Overall Precise(≤{THRESHOLD_DAYS:.1f}d): {overall_within}/{overall_cnt} "
      f"({overall_within/overall_cnt*100:.1f}%)")

print(f"Errors > 5 days: {large_err_count}/{overall_cnt} ({large_err_ratio:.1f}%)")

```

## LSTM

```python
# Sequence modelling configuration
seq_features = [
    'Med_Class', 'ReproductiveCategory', 'Numberpreg', 'Miscarriages', 'Abortions',
    'Breastfeeding', 'CycleWithPeakorNot', 'UnusualBleeding', 'LengthofMenses',
    'LengthofCycle'  # autoregressive input: past cycle length
]
target_col = 'LengthofCycle'
group_col  = 'ClientID'      # group by participant
order_col  = 'CycleNumber'   
W = 3                        # sequence window length
```

```python
def build_sequences(df, features, target, group_col, order_col, window):
    """
    Convert cycle-level records into sliding window sequences per individual.
    Each sequence uses the previous `window` cycles to predict the next one.
    """
    X_list, y_list, g_list = [], [], []

    for gid, gdf in df.sort_values([group_col, order_col]).groupby(group_col):
        gX = gdf[features].values
        gy = gdf[target].values

        if len(gdf) < window + 1:
            continue

        for t in range(window, len(gdf)):
            X_list.append(gX[t-window:t])  # sequence input
            y_list.append(gy[t])           # next cycle target
            g_list.append(gid)

    X = np.stack(X_list)              # shape: (N, window, num_features)
    y = np.array(y_list, dtype=float)
    groups = np.array(g_list)
    return X, y, groups

# Build sequence dataset
X_seq, y_seq, groups_seq = build_sequences(df_new, seq_features, target_col, group_col, order_col, W)
print(X_seq.shape, y_seq.shape, len(np.unique(groups_seq)))
```

```python
from sklearn.model_selection import GroupShuffleSplit

# Split into train / test ensuring each client appears in only one split
gss1 = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss1.split(X_seq, y_seq, groups=groups_seq))

# Further split training set into train / validation (also grouped)
gss2 = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=43)
train_idx2, val_idx = next(gss2.split(X_seq[train_idx], y_seq[train_idx], groups=groups_seq[train_idx]))
train_idx = train_idx[train_idx2]

# Final dataset partitions
X_tr, y_tr = X_seq[train_idx], y_seq[train_idx]
X_va, y_va = X_seq[val_idx],  y_seq[val_idx]
X_te, y_te = X_seq[test_idx], y_seq[test_idx]

print("train/val/test:", X_tr.shape, X_va.shape, X_te.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Number of input features
F = X_tr.shape[-1]

# LSTM model
model = keras.Sequential([
    layers.Input(shape=(W, F)),   # (window, features)
    layers.LSTM(64),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)               # predict next cycle length
])

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss='mae',
    metrics=['mae']
)

# Train with early stopping
history = model.fit(
    X_tr, y_tr,
    validation_data=(X_va, y_va),
    epochs=50,
    batch_size=64,
    verbose=1,
    callbacks=[keras.callbacks.EarlyStopping(patience=8, restore_best_weights=True)]
)
```

```python
# Evaluate on test set
y_pred = model.predict(X_te).ravel()
mae = mean_absolute_error(y_te, y_pred)
r2  = r2_score(y_te, y_pred)

print(f"LSTM Test MAE: {mae:.3f} days  |  R²: {r2:.3f}")
```
